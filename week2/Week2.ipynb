```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Programming Assignment 1 Data New/ExcelFormattedGISTEMPData2CSV.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
latitude_map={
    "NHem": [u'64N-90N', u'44N-64N', u'24N-44N', u'EQU-24N'][::-1],
    "SHem": [u'24S-EQU', u'44S-24S', u'64S-44S', u'90S-64S']
}
latitude_colormap={
    "NHem": "blue",
    "SHem": "red"
}
first_change_year=1920
second_change_year=1970
```

```python
plt.figure(figsize=(15,10))
for y in ["NHem","SHem"]:
    plt.plot(df["Year"],df[y], label=y, color=latitude_colormap[y])
plt.legend(loc="upper center", fontsize=18, ncol=2)
plt.axvline(x=first_change_year, linestyle="--", color="grey")
plt.axvspan(xmin=1880, xmax=first_change_year, facecolor='0.9', alpha=0.5)
plt.axvline(x=second_change_year, linestyle="--", color="grey")
plt.axvspan(xmin=second_change_year, xmax=2014, facecolor='red', alpha=0.1)
plt.annotate("no change", xy=(1890,40), fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.ylabel("Mean temperature",fontsize=20)
```

The plot shows that until 1920's the mean temperature of both north & south hemisphere are almost constant. However, the temperature starts to increase beyond that. Furthermore, 1970 onwards, the rate of increse of temperature is even rapid. The plots also indicate that the northern hemisphere has shown more rapid increase in temperature as compared to southern in the past 15 years.

```python
plt.figure(figsize=(15,10))
for y in ["NHem","SHem"]:
    plt.plot(df["Year"],df[y].rolling(5).mean(), label=y, color=latitude_colormap[y])
plt.legend(loc="upper center", fontsize=18, ncol=2)
plt.axvline(x=first_change_year, linestyle="--", color="grey")
plt.axvspan(xmin=1880, xmax=first_change_year, facecolor='0.9', alpha=0.5)
plt.axvline(x=second_change_year, linestyle="--", color="grey")
plt.axvspan(xmin=second_change_year, xmax=2014, facecolor='red', alpha=0.1)
plt.annotate("no change", xy=(1890,40), fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.ylabel("Mean temperature (rolling average over 5 years)",fontsize=20)
```

The same data is plotted here except on y-axis the rooling avaerage (5 years) is plotted to achive a smoother curve.

```python
plt.figure(figsize=(15,10))
for y in ["NHem","SHem"]:
    plt.plot(df["Year"],df[y].rolling(5).mean(), label=y, color=latitude_colormap[y])
    for i, y1 in enumerate(latitude_map[y]):
        plt.plot(df["Year"],df[y1].rolling(5).mean(), label=None,
                 linewidth=1, color=latitude_colormap[y], alpha=0.1*(i+1))
plt.legend(loc="upper center", fontsize=18, ncol=2)
plt.axvline(x=first_change_year, linestyle="--", color="grey")
plt.axvspan(xmin=1880, xmax=first_change_year, facecolor='0.9', alpha=0.5)
plt.axvline(x=second_change_year, linestyle="--", color="grey")
plt.axvspan(xmin=second_change_year, xmax=2014, facecolor='red', alpha=0.1)
plt.annotate("no change", xy=(1890,40), fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.ylabel("Mean temperature (rolling average over 5 years)",fontsize=20)
```

In the above plot, the temeperature for each latitude zone is also plotted.

```python
plt.figure(figsize=(15,10))
for y in ["NHem","SHem"]:
    plt.plot(df["Year"],df[y], label=y, color=latitude_colormap[y])
    for i, y1 in enumerate(latitude_map[y]):
        plt.plot(df["Year"],df[y1], label=None,
                 linewidth=0.3, color=latitude_colormap[y], alpha=0.2*(i+1))
plt.legend(loc="upper center", fontsize=18, ncol=2)
plt.axvline(x=first_change_year, linestyle="--", color="grey")
plt.axvspan(xmin=1880, xmax=first_change_year, facecolor='0.9', alpha=0.5)
plt.axvline(x=second_change_year, linestyle="--", color="grey")
plt.axvspan(xmin=second_change_year, xmax=2014, facecolor='red', alpha=0.1)
plt.annotate("no change", xy=(1890,40), fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.ylabel("Mean temperature",fontsize=20)
```
